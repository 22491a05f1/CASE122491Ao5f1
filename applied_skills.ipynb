```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
ramya=pd.read_csv('energy.csv')
print(ramya)
srh=ramya[ramya['State']=='Andhra']
rcb=ramya[ramya['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='red')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='yellow')



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
sumanth=pd.read_csv('energy.csv')
print(sumanth)
#line chart
srh=sumanth[sumanth['State']=='Andhra']
rcb=sumanth[sumanth['State']=='Karnataka']
fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])

srh=sumanth[sumanth['State']=='Telengana']
rcb=sumanth[sumanth['State']=='Tripura']
fig,diag=plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'])
diag[1].bar(rcb['Date'],rcb['Consumption'])

cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]
srh=sumanth[sumanth['State']=='Kerala']['Consumption'].sum()
rcb=sumanth[sumanth['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]

fig,diag=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```

```python
import pandas as cr
import matplotlib.pyplot as lr
import seaborn as cse

ntr=cr.read_csv('heart.csv')
print(ntr)

cse.scatterplot(x='age',y='bp',data=ntr,hue='gender')
lr.title('heart attach')
lr.xlabel('age')
lr.ylabel('cholestrol')
lr.show()

#line chart
cse.lineplot(x='age',y='cholestrol',data=ntr)
lr.show()

#box chart
cse.boxplot(x='target',y='value',data=ntr)
lr.show()

#countplot
color=['red','yellow']
cse.countplot(x='target',data=ntr,palette=color)
lr.show()





```
